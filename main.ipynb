```python
sepal_length=list(iris_data['sepal length (cm)'])
sepal_width=list(iris_data['sepal width (cm)'])
petal_length=list(iris_data['petal length (cm)'])
petal_width=list(iris_data['petal width (cm)'])
target_class=list(iris_data['species'])
```

```python
#encoding features
# Import LabelEncoder
from sklearn import preprocessing
#creating labelEncoder
le = preprocessing.LabelEncoder()
# Converting string labels into numbers.
target_class_encoded=le.fit_transform(target_class)
print('target_class',target_class_encoded)
```

```python
features=zip(sepal_length,sepal_width,petal_length,petal_width)
features=list(features)
#target_class_encoded=list(target_class_encoded)
print(features)
```

```python
#Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()#Create a Gaussian Classifier
model.fit(features,target_class_encoded)# Train the model using the training sets
#Predict Output
predicted= model.predict([[5.6,2.3,1.4,5.3]])#sepal length, sepal width, petal length, petal width
print("Predicted Value:", predicted)
```

```python
# Importing train_test_split function
from sklearn.model_selection import train_test_split
# Split dataset into training set and test set  70% training and 30% test
X_train, X_test, y_train, y_test = train_test_split(features,target_class_encoded, test_size=0.3,random_state=109) 
```

```python
gnb = GaussianNB()#Create a Gaussian Classifier
gnb.fit(X_train, y_train)#Train the model using the training sets
y_pred = gnb.predict(X_test)#Predict the response for test dataset
```

```python
#accuracy calculation
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
